```python
import pandas as pd
import numpy as np


df = pd.read_csv("student_exam_scores_12_13.csv")
df.head()
```

```python

# ------------------------------------------------------------
# 12. Linear Regression from scratch (1 feature: hours_studied)
# ------------------------------------------------------------

# Extract features and target
x = df['hours_studied'].values
y = df['exam_score'].values
n = len(x)

# Hyperparameters
alpha = 0.01
epochs = 2000

# Initialize parameters
m = 0.0
c = 0.0

# Gradient Descent
for epoch in range(epochs):
    y_pred = m * x + c
    error = y_pred - y
    dm = (1/n) * np.sum(error * x)
    dc = (1/n) * np.sum(error)
    m = m - alpha * dm
    c = c - alpha * dc

# Predictions
y_pred = m * x + c

# Evaluation Metrics
mse = np.mean((y_pred - y)**2)
r2 = 1 - (np.sum((y - y_pred)**2) / np.sum((y - np.mean(y))**2))

print("ðŸ“˜ Linear Regression (1 Feature - hours_studied)")
print(f"Equation: exam_score = {m:.3f} * hours_studied + {c:.3f}")
print(f"MSE: {mse:.3f}")
print(f"RÂ² Score: {r2:.3f}")
```
